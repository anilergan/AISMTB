### How can I get Bitcoin’s historical price data in CSV format?

```python
# valid intervals -> 1m, 3m, 5m, 15m, 30m, 1h, 2h, 4h, 6h, 8h, 12h, 1d, 3d, 1w, 1M
```

```python
timestamp = client._get_earliest_valid_timestamp('BTCUSDT', '1h')
timestamp
```

```python
from datetime import datetime

# Veri almak için gerekli parametreleri belirtin.
symbol = "BTCUSDT"
interval = Client.KLINE_INTERVAL_1HOUR  # 1 saatlik aralık

# Tarih aralığını belirtin (2023.09.01 00:00:00'dan 2023.09.08 00:00:00'a kadar).
start_date = "2023-09-01"
end_date = "2023-09-08"

# Geçmiş kapanış fiyatlarını alın.
historical_klines = client.get_historical_klines(symbol, interval, start_date, end_date)

```

```python
historical_klines
```

```python
import pandas as pd
# Örnek veri
data = historical_klines

columns = [
    'Timestamp', 'Open Price', 'High Price', 'Low Price', 'Close Price', 'Trade Volume',
    'Close Timestamp', 'Taker Buy Base Asset Volume', 'Taker Buy Base Asset Volume Timestamp',
    'Ignore1', 'Ignore2', 'Ignore3'
]

# Verileri DataFrame'e dönüştürme
df = pd.DataFrame(data, columns=columns)

# İstenen sütunları seçme
selected_columns = ['Timestamp', 'Open Price', 'Close Price', 'High Price', 'Low Price']
df = df[selected_columns]

# Unix zaman damgasını datetime formatına dönüştürme
df['Timestamp'] = pd.to_datetime(df['Timestamp'], unit='ms')

# Sütun adlarını yeniden adlandırma
df.columns = ['Date', 'Açılış Fiyatı', 'Kapanış Fiyatı', 'En Yüksek Fiyat', 'En Düşük Fiyat']

df
```
